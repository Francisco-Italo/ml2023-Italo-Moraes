```python
import pandas as pd

year = 2013
df = pd.read_excel('/content/drive/MyDrive/MDIP_2013_2023.xlsx', sheet_name=None)

union_dfs = df[str(year)]

year = year+1
while year < 2024:
  union_dfs = pd.concat([union_dfs, df[str(year)]], ignore_index=True)
  year = year+1

union_dfs
```

```python
union_dfs.fillna('nao espec.', inplace=True)

union_dfs
```

```python
union_dfs.columns = [col.replace('.', '').replace('Ç', 'C').replace('Ô', 'O').replace('É', 'E').replace('Ê', 'E').replace('Ã', 'A').replace('Á', 'A').replace('Í', 'I') for col in union_dfs.columns]

union_dfs
```

```python
union_dfs = union_dfs.drop('NATUREZA_APURADA', axis=1)
```

```python
union_dfs
```

```python
union_dfs = union_dfs.loc[union_dfs['MUNICIPIO_CIRCUNSCRICAO'] == 'S.PAULO']

union_dfs
```

```python
for i in union_dfs.columns:
  union_dfs[i]=union_dfs[i].astype(str).str.lower()

union_dfs
```

```python
import matplotlib.pyplot as plt

years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
counts = []
for i in years:

  filter = union_dfs['ANO ESTATISTICA'] == str(i)

  counts.append(len([x for x in filter if x == True]))

plt.title("N° de ocorrências por ano"), plt.xlabel("Ano"), plt.ylabel("N° de ocorrências")
plt.bar(years, counts)
```

```python
skin_color = ['preta', 'parda', 'branca']
counts = []
for i in skin_color:

  filter = union_dfs['COR_PELE'] == i

  counts.append(len([x for x in filter if x == True]))

print(counts)
plt.title("N° de ocorrências por cor da pele"), plt.xlabel("Cor da pele"), plt.ylabel("N° de ocorrências")
plt.bar(skin_color, counts)
```
