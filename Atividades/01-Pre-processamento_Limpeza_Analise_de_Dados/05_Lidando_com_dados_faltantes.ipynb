# 5 - Lidando com dados faltantes

## Identificando valores faltantes em dados tabulares
Vamos iniciar nosso estudo identificando dados faltantes em tabelas.

```python
import pandas as pd
# NaN -> Dado nulo
df = pd.read_csv('/content/missing.csv')
df
```

```python
# Retorna a soma dos valores nulos por coluna do dataframe
df.isnull().sum()
```

```python
# Cria um array numpy com todos os valores do DataFrame
df.values
```

## Eliminando amostras ou *features* com valores ausentes

```python
# Remove colunas que contém valores faltantes no eixo desejado
# 0 -> linha | 1 -> coluna

df.dropna(axis=0)
```

```python
# Removendo colunas que contém valores faltantes
df.dropna(axis=1)
```

```python
df.dropna(axis=0)
```

```python
# "Drop" apenas nas linhas onde todas as colunas são NaN

df.dropna(how='all')
```

```python
# Removendo linhas que possuem menos de 3 valores reais

df.dropna(thresh=4)
```

```python
df.dropna(thresh=3)
```

```python
# Remove apenas linhas onde NaN aparece em uma coluna específica

df.dropna(subset=['C'])
```

```python
df.dropna(subset=['D'])
```

```python
df.dropna(subset=['A'])
```
